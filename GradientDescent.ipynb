```python
#Step 1: Iitialize the slope and the Intercept values
#Step 2: Initializr the Learning Rate and the iterations
#Step 3: Loop(iterations):
#Step 4: Y_prediction = m*X+c
#Step 5: PD value of Slope and Intercept 
        #[Partial Differential Calculas Slope = -2/n*sum(X-(Y-Y_Pred))] 
        #[Partial Differential Calculas Intercept = -2/n*sum(Y-Y_Pred)] 
#Step 6: Calculate the new Slope and Intercept values [New_Slope = Old_slope - LearningRate*Delta_Slope]
#Step 7: Print the Final Values of M and C 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Laxminarayen/Inceptez-Batch-17/main/Day%205%20Linear%20Regression%20Hands%20on/data.csv")
```

```python
X= df.iloc[:,0]
Y= df.iloc[:,1]
plt.scatter(X,Y)
#plt.plot([min(X),max(X)],[min(Y),max(Y)],color = 'red')
plt.show()
```

```python
m=0
c=0
Y_pred = m*X+c
plt.scatter(X,Y)
plt.plot([min(X),max(X)],[min(Y_pred),max(Y_pred)],color = 'red')
plt.show()
```

```python
L = 0.0001
iter_ = 1000

n = float(len(X))

#For loop - Performing Gradient Descent 
for i in range(iter_):
    Y_pred = (m*X)+c #The current prediction value 
    D_m = (-2/n)*sum(X * (Y - Y_pred)) #Delta Slope
    D_c = (-2/n)*sum(Y - Y_pred) #Delta Intercept
    
    m = m - (L*D_m) 
    c = c - (L*D_c)
     
    print("The Slope and intercept for iteration:",i+1)
    print("Slope: ",m)
    print("Intercept: ",c)

print("After 1000 iterations values of slope: %0.2f and intercept: %0.2f " %(m,c))
```
